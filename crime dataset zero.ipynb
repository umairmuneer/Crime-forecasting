```python
from calendar import monthrange
import datetime

def rangetime(NoOfYear):
    coount=0
    cre=[]
    for yr in NoOfYear:
        print(yr)
        for mon in range(1, 13):
            print(mon)
            day=monthrange(yr, int(mon))
            day=list(day)
            for dy in range(1,day[1]+1):
                for hu in range(24):
                    fff1=datetime.datetime(yr, mon, dy,hu,0)
                    cre.append(fff1)
                    coount=coount+1
    return cre         
```
